```python
dataset=pd.read_csv("placement.csv")
dataset.head(3)
```

```python
plt.figure(figsize=(4,5))
sns.scatterplot(x="cgpa",y="package",data=dataset)
plt.show()
```

```python
x = dataset.iloc[:, :-1]
y = dataset["package"]

# Convert the continuous 'package' variable into a categorical variable
median_package = y.median()
y = (y > median_package).astype(int)
```

```python
from sklearn.model_selection import train_test_split
```

```python

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import Perceptron
```

```python
pr=Perceptron()
pr.fit(x_train,y_train)
```

```python
pr.score(x_test,y_test)*100,pr.score(x_train,y_train)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=pr)
plt.show()
```
